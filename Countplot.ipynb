```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('practice.csv')
df
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='sex',data=df)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='island',data=df,hue='sex',palette='cool')
```
